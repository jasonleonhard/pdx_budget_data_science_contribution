```python
ATAT000001 = df[df.org == "ATAT000001"]
ATAT000001
```

```python
ATAT000001_numeric = pd.to_numeric(ATAT000001['fund'])

plt.figure()
plt.hist(ATAT000001_numeric, log=True, bins=2)

plt.xlabel('count')
plt.ylabel('$')
plt.title(r'$\mathrm{Histogram\ of\ Count\ by\ dollar\ amount}\ $')

plt.show()
```

#### Alternative way to plot.

```python
ATAT000001_numeric.count() # 842 in size
x = ATAT000001_numeric
n, bins, patches = plt.hist(x, 2, facecolor='green', alpha=0.75, log=True)

plt.xlabel('count')
plt.ylabel('$')
plt.title(r'$\mathrm{Histogram\ of\ Count\ by\ dollar\ amount}\ $')

plt.show()
```
