#### I was given a messy tab seperated data file that will need some cleaning up prior to analysis. First, I import the necessary libraries and load in the data.

#### The dataframe will not accept large paragrapsh of text, for now I have removed columns with lots of text and saved a new version of the data file to allow for transparency of thought process. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
fname = "../data/Funded_program_attributes_mod2.txt"
```

```python
df0 = pd.read_csv(fname, sep='\t', header=0)
df0.count() 
```

#### The largest column count is 6368, but I must have a dollar value, so later, I look to column 'value4' which is shown to have 3860 line items.

#### Now I view the dataframe.

```python
df0
```

#### There are too many columns, many with only "NaN" meaning not a number, so to make the dataframe more manageable for now I will only select all data in a choice few columns.

```python
# only have headers in dataframe you want
df1 = df0.loc[:,['org','name','tbl1','value1','value4']]
df1.head(20)
```

#### In fact, since I really want to just clean up the data enough to analyze dollar amounts, I can ignore all columns except for "value4" for now.

```python
v4 = df1.loc[:,['value4']]
v4.head()
```

#### Many values are NaN, in this column and in others as well, I currently can do nothing with these NaN so I will remove them.

```python
v4 = v4.dropna()
v4.tail(40)
# type(v4) # pandas.core.frame.DataFrame
```

#### I now see the string "Ongoing", it will need to be removed for now.

```python
v4 = v4[v4.value4 != "Ongoing"]
v4.tail(40)
```

#### One might assume the values viewed above were integers, but they would be incorrect.  I will need to remove all of the:
#### 1. "," commas
#### 2. "$" dollar signs and
#### 3.  "," empty spaces

```python
v4.dtypes # value4    object
          # dtype: object

v4['value4'] = v4.value4.apply(lambda b: b.replace(',', ''))
v4['value4'] = v4.value4.apply(lambda b: b.replace('$', ''))
v4['value4'] = v4.value4.apply(lambda b: b.replace(' ', ''))
```

```python
type(v4['value4']) # pandas.core.series.Series
type(v4)           # pandas.core.frame.DataFrame
v4.dtypes          # value4 object
                   # dtype: object
```

#### I still do not have an interger or float value to operate on, I will cast the type now.

```python
v4_num = pd.to_numeric(v4['value4'])
v4_num
type(v4_num)     # pandas.core.series.Series
# alternatively use shift tab when above a variable or function

v4_num.dtypes    # dtype('float64')
```

#### Finally, I have a float which will work, I could have used an integer instead but this will work fine. I will now show how I could create columns/series that would store all values above 100,000,000.

```python
above_1e8 = v4_num[(v4_num > 1e8)] # or use
above_1e8.count() # 9
above_1e8

# 948     537148518.0 # max 
# 5501    116304262.0 # min
```

#### To plot a distribution and check if it is normal, such as with a gaussian distribution, I import the math llibrary and create a list with log base 10 for all numeric values to later plot a histogram.

```python
import math
logv4_num = [math.log10(i) for i in v4_num]
logv4_num
```

#### Plotting a histogram we can see the distribution by eye for now.
